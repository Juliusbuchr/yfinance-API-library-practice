# yfinance library for financial data

I am getting tremendously tired of CSV-files so I want to check out how easy it is to get data with the yfinance library.

```python
# Importing prerequisites

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 

# And the most important; installing and importing yfinance

# !pip install yfinance (if not already installed)

import yfinance as yf 
```

```python
# Basic example; how to import data from a ticker

msft = yf.Ticker("MSFT")
y = msft.info
msft.info
```

```python
# I must say that I am already getting excited!
# There is a lot of useful data here

# Getting historical data

hist = msft.history(period="max")
hist
```

```python
# This already looks like a pandas data frame and I think it might be

# We can also view corporate actions

msft.actions 
```

```python
# We can even get information on major share holders

msft.major_holders
```

```python
# And further view who these institutional holders are

msft.institutional_holders
```

```python
# Viewing analyst recommendations

msft.recommendations
```

```python
# We can see a calendar for upcoming events 

msft.calendar
```

```python
# We can get the options prices

msft.options
```

```python
# Plotting historical data (all subplots)

hist.plot(kind="line", subplots=True)
```

```python
# Lets have some fun now. Lets see again what data is available (refer to info box)
# We can do something fun e.g. calculate short percentage of floating shares
# Remember we defined y as a dictionary with the information

msft.info

short = y["sharesShort"]
floating = y["floatShares"]

short_pct = short/floating*100
short_pct
```

```python
# Plotting the share price

hist["Close"].plot()
plt.ylabel("Share price")
plt.title("MSFT")
```
